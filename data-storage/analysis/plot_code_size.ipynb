## Code size

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

sizes = {
    'AES-GCM': 906,
    'ASCON': 902,
    'Masked ASCON': 963,
    'NONE': 885
}

# Subtract 'none'
base = sizes['NONE']
code_sizes_kb = {
    k: v - base for k, v in sizes.items() if k != 'NONE'
}

# Order and color palette
labels = ['AES-GCM', 'ASCON', 'Masked ASCON']
values = [code_sizes_kb[k] for k in labels]
colors = ['#2ca02c', '#1f77b4', '#ff7f0e']  # green, blue, orange

# Plot
plt.figure(figsize=(6, 4))
bars = plt.bar(labels, values, color=colors)
plt.ylabel("Code Size [kb]")
plt.title("Encryption Method Code Size")
plt.grid(axis='y')

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 1, f'{height} kb', 
             ha='center', va='bottom', fontsize=9)

plt.ylim(0, 90)

plt.tight_layout()
path = "../figures/"
plt.savefig(path + "code_size.png", dpi=300)
plt.show()
```
